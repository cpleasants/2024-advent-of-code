```python
class Day19:
    def __init__(self, fname):
        with open(fname) as f:
            self.patterns = []
            for i, line in enumerate(f.readlines()):
                if i == 0:
                    self.towels = [t.strip() for t in line.strip().split(',')]
                elif line.strip() == '':
                    continue
                else:
                    self.patterns.append(line.strip())

    def find_possible_matches(self, pattern):
        possible_matches = [t for t in self.towels if t in pattern]
        # It is redundant to include matches which can be made up of other matches
        # But also track the ones that were removed to be able to count the alternatives later
        track = defaultdict(list)
        cont = True
        while cont:
            removed = 0
            for match1 in possible_matches:
                for match2 in possible_matches:
                    if match1 + match2 in possible_matches:
                        track[tuple([match1, match2])] = match1 + match2
                        possible_matches.remove(match1 + match2)
                        removed += 1
            if removed == 0:
                cont = False
        return possible_matches, track

    def t(self, pattern, possible_matches, lists = [[]]):
        new_lists = []
        for ls in lists:
            remaining_pattern = pattern
            for item in ls:
                remaining_pattern = remaining_pattern.split(item, 1)[1]
                if remaining_pattern == '':
                    new_lists.append(ls)
            for match in possible_matches:
                if match in remaining_pattern:
                    if remaining_pattern.split(match)[0] == '':
                        new_lists.append(ls + [match])
        if len(new_lists) == 0:
            return False
        elif new_lists == lists:
            return lists
        else:
            return self.t(pattern, possible_matches, new_lists)

    def count_all_solutions(self, solutions, tracked_eliminations):
        # for solution in solutions:
        #     for elimination in tracked_eliminations:
        #         if any(solution[i:i + 2] == list(elimination) for i in range(len(solution) - 1)):
                    
        return 1
        


    def solve(self):
        count_true = 0
        count_solutions = 0
        # print(len(self.patterns))
        for i, pattern in enumerate(self.patterns): 
            if i % 20 == 0:
                print(i)
            possible_matches, tracked_eliminations = self.find_possible_matches(pattern)
            print(tracked_eliminations)
            solutions = self.t(pattern, possible_matches)
            if solutions is False:
                continue
            else:
                count_true += 1
                count_solutions += self.count_all_solutions(solutions, tracked_eliminations)

            # count_true += self.t(pattern, possible_matches)
        return count_true, count_solutions
                
        
                    
```
